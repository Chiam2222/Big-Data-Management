## Overview

This notebook shows you how to load data using Spark SQL from CSV files that have been bzip2 archived.

### Step 1: Reading the data

Now that we have uploaded the movies.csv.bz2, tags.csv.bz2 and ratings.csv.bz2 files, we can create a DataFrame. You'll notice that we use an *option* to specify that we'd like to infer the schema from the file. We can also explicitly set this to a particular schema if we have one already.

```
destination = "/FileStore/tables/Assignment_Data/"

movies_DF = (spark.read
  .option("sep", ",")
  .option("header", True)
  .option("inferSchema", True)
  .csv(destination + "movies_csv.bz2"))

movies_DF.printSchema()
```

```
movies_DF.display()
```

```
destination = "/FileStore/tables/Assignment_Data/"

ratings_DF = (spark.read
  .option("sep", ",")
  .option("header", True)
  .option("inferSchema", True)
  .csv(destination + "ratings_csv.bz2"))

ratings_DF.printSchema()
```

```
ratings_DF.display()
```

```
destination = "/FileStore/tables/Assignment_Data/"

tags_DF = (spark.read
  .option("sep", ",")
  .option("header", True)
  .option("inferSchema", True)
  .csv(destination + "tags_csv.bz2"))

tags_DF.printSchema()
```

```
tags_DF.display()
```

### Step 2: Querying the data

Now that we created our DataFrame. We can query it.

```
ratings_DF.join(movies_DF, ratings_DF.movieId == movies_DF.movieId, 'inner').select(
  ratings_DF.userId, movies_DF.title, ratings_DF.rating).limit(100).display()
```

```
tags_DF.join(movies_DF, tags_DF.movieId == movies_DF.movieId, 'outer').select(movies_DF.title, tags_DF.tag).filter(tags_DF.tag == 'comedy').limit(100).display()
```

## Part 1 Question 1
Extract the year of release of the movie from the title and insert a new column named 'yearOfRelease'.

```
from pyspark.sql.functions import *

movies_DF2 = movies_DF.withColumn("yearOfRelease", movies_DF.title.substr(-5,4))
movies_DF3 = movies_DF2.withColumn("yearOfRelease", concat_ws("", col("yearOfRelease")))
movies_DF3.display()
```

## Question 2
Create a visualization showing the number of movies broken down by the year of release for year 1990 onwards.

```
from pyspark.sql.functions import *

movies_DF4 = movies_DF3.withColumn("yearOfRelease",movies_DF3.yearOfRelease.cast('int'))
# movies_DF5 = movies_DF4.orderBy(desc("yearOfRelease"))
movies_DF4.select("title", "yearOfRelease").display()
```

```
import matplotlib.pyplot as plt
from pyspark.sql.functions import col, sum

movies_DF6 = movies_DF5.groupBy("yearOfRelease").agg(count(col("movieId").cast("Integer")).alias("numOfMovies")).orderBy(col("yearOfRelease"))
movies_DF7 = movies_DF6.filter((movies_DF5['yearOfRelease'] >= 1990))
display(movies_DF7)
```

```
import matplotlib.pyplot as plt
import numpy as np

yearOfReleaseDF = movies_DF7.toPandas()
plt.figure()
plt.bar(yearOfReleaseDF.yearOfRelease, yearOfReleaseDF.numOfMovies)

plt.xlabel("yearOfRelease")
plt.ylabel("numOfMovies")

x_ticks = np.arange(1990, 2020, 2)
plt.xticks(x_ticks, rotation=270)

plt.grid(axis = 'y')
plt.show()
```

## Question 4
Create a DataFrame or RDD that shows the different generations of movies and their average rating. The expected output of the table is shown below.

```
newDF = movies_DF4.withColumn("year_range",
                     when((movies_DF4.yearOfRelease >= 1960) & (movies_DF4.yearOfRelease <= 1969), lit("1960-1969")).
                     when((movies_DF4.yearOfRelease >= 1970) & (movies_DF4.yearOfRelease <= 1979), lit("1970-1979")).
                     when((movies_DF4.yearOfRelease >= 1980) & (movies_DF4.yearOfRelease <= 1989), lit("1980-1989")).
                     when((movies_DF4.yearOfRelease >= 1990) & (movies_DF4.yearOfRelease <= 1999), lit("1990-1999")).
                     when((movies_DF4.yearOfRelease <= 1959), lit("1959 and earlier")).
                     when((movies_DF4.yearOfRelease >= 2000), lit("2000 onwards"))
                     .otherwise(lit(" ")))
# newDF.filter(newDF.yearOfRelease.isNull()).display()
newDF.select("yearOfRelease", "year_range").display()
```

```
newDF2 = ratings_DF.join(newDF, 'movieId').select(newDF.year_range, ratings_DF.rating)
newDF2.display()
```

```
newDF2.groupBy("year_range") \
    .agg(avg("rating").alias("avg(rating)"), \
     ) \
    .show(truncate=False)
```
